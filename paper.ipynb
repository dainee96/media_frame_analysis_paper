```python
import pandas as pd
df = pd.read_csv('전체언론사_20200201-20200831.csv', error_bad_lines = False)
df
```

```python
del df['Unnamed: 0']
```

```python
df
```

```python
df = df.sort_values(by=['언론사'],ascending=True)
df
```

```python
set(df['언론사'])
```

```python
key = df['키워드']
key
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word',stop_words=None,min_df=20,max_features=3000)
vectors = tfidf.fit_transform(key)
feature_names = tfidf.get_feature_names()
vectors.shape
```

```python
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
df.to_csv('언론사들_tfidf.csv',encoding = 'utf-8-sig',header=True)
```
