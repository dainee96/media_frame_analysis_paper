```python
ls = []
for i, standard in enumerate(lst):
    stan = set()
    tar = set()
    for z in standard:
        stan.add(z)
        
    for j, target in enumerate(lst):
        if i == j:
            continue
        for zz in target:
            tar.add(zz)
    etc = stan-tar
    ls.append(list(etc))
ls
```
