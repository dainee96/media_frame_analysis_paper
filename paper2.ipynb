```python
import pandas as pd
ar = pd.read_excel('코로나_20200201-20200831.xlsx')
ar.head()
```

```python
import numpy as np
lstt = np.array(ar)
lst = np.transpose(lstt)
lst
```

```python
lst.shape
```

```python
ls = []
for i, standard in enumerate(lst):
    stan = set()
    tar = set()
    for z in standard:
        stan.add(z)
        
    for j, target in enumerate(lst):
        if i == j:
            continue
        for zz in target:
            tar.add(zz)
    etc = stan-tar
    ls.append(list(etc))
ls
```

```python
n = []
m = []
for i in range(24):
    for j in range(24):
        it = np.intersect1d(lst[i],lst[j])
        m = np.append(m, set(it))
        n = np.append(n, len(it))
matrix = np.array(m)
num = np.array(n,dtype=int)
print(matrix)
print(num)
```

```python
nm = num.reshape(24,24)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(nm)
x_pca = pca.transform(nm)
print(x_pca.shape)
print(x_pca)
```
